```python
import pandas as pd
from datetime import datetime
```

```python
def date_string(x):
    return datetime.utcfromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S')
```

```python
df = pd.read_csv("scooter_info2.csv", header=None)
```

```python
df[6] = df[6].map(date_string)
```

```python
df.to_csv("scooter_parsed.csv")
```
